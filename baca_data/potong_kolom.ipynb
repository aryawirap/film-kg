# Potong Kolom Dataset untuk Knowledge Graph Film
Hanya ambil kolom yang diperlukan dari setiap CSV, lalu simpan ke folder `dataset/`.

```python
import pandas as pd
import os

DATA_DIR   = os.path.join('..', 'database')
OUTPUT_DIR = os.path.join('..', 'dataset')
os.makedirs(OUTPUT_DIR, exist_ok=True)
```

## 1. movies_metadata.csv
Kolom yang dipakai: `id`, `title`, `genres`, `release_date`, `revenue`

```python
df_movies_raw = pd.read_csv(os.path.join(DATA_DIR, 'movies_metadata.csv'), low_memory=False)
print('Semua kolom:', list(df_movies_raw.columns))
```

```python
COLS_MOVIES = ['id', 'title', 'genres', 'release_date', 'revenue']

df_movies = df_movies_raw[COLS_MOVIES].copy()

# Hapus baris dengan id tidak valid (ada beberapa baris corrupt)
df_movies = df_movies[pd.to_numeric(df_movies['id'], errors='coerce').notna()]
df_movies['id'] = df_movies['id'].astype(int)

print(f'Shape: {df_movies.shape}')
df_movies.head()
```

```python
df_movies.to_csv(os.path.join(OUTPUT_DIR, 'movies.csv'), index=False)
print('Tersimpan: dataset/movies.csv')
```

## 2. credits.csv
Kolom yang dipakai: `id`, `cast`, `crew`

```python
df_credits_raw = pd.read_csv(os.path.join(DATA_DIR, 'credits.csv'))
print('Semua kolom:', list(df_credits_raw.columns))
```

```python
COLS_CREDITS = ['id', 'cast', 'crew']

df_credits = df_credits_raw[COLS_CREDITS].copy()

print(f'Shape: {df_credits.shape}')
df_credits.head()
```

```python
df_credits.to_csv(os.path.join(OUTPUT_DIR, 'credits.csv'), index=False)
print('Tersimpan: dataset/credits.csv')
```

## 3. keywords.csv
Kolom yang dipakai: `id`, `keywords`

```python
df_keywords_raw = pd.read_csv(os.path.join(DATA_DIR, 'keywords.csv'))
print('Semua kolom:', list(df_keywords_raw.columns))
```

```python
COLS_KEYWORDS = ['id', 'keywords']

df_keywords = df_keywords_raw[COLS_KEYWORDS].copy()

print(f'Shape: {df_keywords.shape}')
df_keywords.head()
```

```python
df_keywords.to_csv(os.path.join(OUTPUT_DIR, 'keywords.csv'), index=False)
print('Tersimpan: dataset/keywords.csv')
```

## Ringkasan Output

```python
summary = {
    'movies.csv':   df_movies,
    'credits.csv':  df_credits,
    'keywords.csv': df_keywords,
}

print(f'{"File":<20} {"Shape":<15} Kolom')
print('-' * 60)
for name, df in summary.items():
    print(f'{name:<20} {str(df.shape):<15} {list(df.columns)}')
```
